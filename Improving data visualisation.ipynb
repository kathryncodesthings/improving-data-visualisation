```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("CNBC income data.csv")

display(df)

display(df.info())
```

```python
df_pivot = df.pivot('Revenue source','Year','CAD$')

display(df_pivot)

df_pivot.plot(kind='bar', stacked=True)

# Just add a title and rotate the x-axis labels to be horizontal.
plt.title('CNBC Revenue, 2017-18 and 2018-19')
plt.xticks(rotation=0, ha='center')
```
